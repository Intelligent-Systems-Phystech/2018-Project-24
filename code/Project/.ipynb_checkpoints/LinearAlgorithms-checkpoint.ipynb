```python
import numpy as np
import torchvision as tv
import matplotlib.pyplot as plt
import torch
import os
import torch.nn as nn
import BasicFunctions as bf

from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
#from torch.nn import functional as F 
```

```python
os.chdir('C:\\Users\\Yaroslav\\YandexDisk\\Pr\\Py\\ML\\DataCreation\\output')
X = np.load('X_base_line_ones30.npz')
y = np.load('Y_base_line30.npz')
name = 'city'
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[name], y[name])

train_loader = DataLoader(
    dataset=list(zip(X_train, y_train)), 
    batch_size=100, 
    shuffle=True)

test_loader = DataLoader(
    dataset=list(zip(X_test, y_test)), 
    batch_size=10, 
    shuffle=True)
```

```python
class basicNet(nn.Module):
    def __init__(self):
        super(basicNet, self).__init__()
    
        #self.net = nn.Sequential()
        self.lin1 = nn.Linear(100, 200)
        self.lin2 = nn.Linear(200, 300)
        self.lin3 = nn.Linear(300, 400)
        self.lin4 = nn.Linear(400, 500)
        self.lin5 = nn.Linear(500, 600)
        self.lin6 = nn.Linear(600, 700)
        self.lin7 = nn.Linear(700, 800)
        self.lin8 = nn.Linear(800, 900)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.lin1(x))
        x = self.relu(self.lin2(x))
        x = self.relu(self.lin3(x))
        x = self.relu(self.lin4(x))
        x = self.relu(self.lin5(x))
        x = self.relu(self.lin6(x))
        x = self.relu(self.lin7(x))
        x = self.relu(self.lin8(x))
        return x
```

```python
class basicNet_pool(nn.Module):
    def __init__(self):
        super(basicNet_pool, self).__init__()
    
        #self.net = nn.Sequential()
        self.lin1 = nn.Linear(100, 400)
        self.lin2 = nn.Linear(400, 800)
        self.lin3 = nn.Linear(800, 1200)
        self.lin4 = nn.Linear(1200, 1600)
        self.lin5 = nn.Linear(1600, 2000)
        self.lin6 = nn.Linear(2000, 2400)
        self.lin7 = nn.Linear(2400, 2800)
        self.lin8 = nn.Linear(2800, 3200)
        self.lin9 = nn.Linear(3200, 3600)
        self.relu = nn.ReLU()
        self.pad = nn.ReplicationPad2d(1)
        self.avgpool = nn.AvgPool2d((2, 2), stride=(2, 2))

    def forward(self, x):
        x = self.relu(self.lin1(x))
        x = self.relu(self.lin2(x))
        x = self.relu(self.lin3(x))
        x = self.relu(self.lin4(x))
        x = self.relu(self.lin5(x))
        x = self.relu(self.lin6(x))
        x = self.relu(self.lin7(x))
        x = self.relu(self.lin8(x))
        x = self.relu(self.lin9(x))
        x = self.avgpool(x.reshape((60, 60)))
        
        return x.reshape(900)
```

```python
cuda_net = basicNet_pool()
cuda_net = cuda_net.cuda()

learning_rate = 0.001
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(cuda_net.parameters(), lr=learning_rate)
```

```python
train_cuda_MSE, test_cuda_MSE = bf.net_cuda_training(cuda_net, train_loader, test_loader, criterion, learning_rate, 200)
```

```python
plt.plot(train_cuda_MSE, label='train')
plt.plot(test_cuda_MSE, label= 'test')
plt.legend()
```
